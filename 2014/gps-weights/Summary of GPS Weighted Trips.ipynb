```python
import pandas as pd
```

```python
# Load GPS-weighted trip records
trip_wt = pd.read_csv(r'R:\SoundCastDocuments\2014Estimation\Files_From_Mark_2014\gps_weights_11_3_16\formatted\skims_attached\tripP14_w.dat')

# Load non-GPS-weighted trips
trip_nwt = pd.read_csv(r'R:\SoundCastDocuments\2014Estimation\Files_From_Mark_2014\new_weights_10_28_16\skims_attached\tripP14.dat', delim_whitespace=True)
```

```python
# trip_wt.groupby('travdist').sum()['trexpfac']
trip_wt['travdist_bin'] = trip_wt['travdist'].apply(lambda row: str(row).split('.')[0]).astype('int')
trip_wt = trip_wt[trip_wt['travdist_bin'] <= 30]
```

```python
trip_wt.groupby(['dpurp','travdist_bin']).sum()['trexpfac'].to_clipboard()
```

```python
trip_wt.groupby(['mode','travdist_bin']).sum()['trexpfac']
```

```python
# trip_wt.groupby('travdist').sum()['trexpfac']
trip_nwt['travdist_bin'] = trip_nwt['travdist'].apply(lambda row: str(row).split('.')[0]).astype('int')
trip_nwt = trip_nwt[trip_nwt['travdist_bin'] <= 30]
```

```python
trip_nwt.groupby(['dpurp','travdist_bin']).sum()['trexpfac'].to_clipboard()
```
