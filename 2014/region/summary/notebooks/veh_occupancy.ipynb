```python
import numpy as np
import pandas as pd
import load_data as data
```

```python
# person trips by car
person_trips_auto = data.trip.query("mode == 1 or mode == 2 or mode == 3 or mode == 4 or mode == 5")
auto_trips = person_trips_auto.query("driver == 1")
```

```python
# avg vehicle occupancy (all person trips by )
veh_occ = person_trips_auto['expwt_fina'].sum()/auto_trips['expwt_fina'].sum()
```

```python
# Vehicle occupancy by county

df = person_trips_auto.groupby('dcnty').sum()['expwt_2']/auto_trips.groupby('dcnty').sum()['expwt_2']
df.to_clipboard()
```
